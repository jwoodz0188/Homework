```python
import string
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from textblob import TextBlob
from textblob.taggers import NLTKTagger
from nltk.probability import FreqDist
import plotly.express as px
import numpy as np;
from PIL import Image;
import nltk
nltk.download('punkt')
```

```python
!pip install requests

import requests

target_url ='http://www.gutenberg.org/files/2265/2265.txt'

response = requests.get(target_url)

data = response.text
```

```python
blob = TextBlob(data)
blob.word_counts['i']

frequency = pd.DataFrame.from_dict(blob.word_counts, orient='index', columns=['count'])

px.bar(frequency.sort_values(by='count', ascending=False)[:20])
```

```python
import imageio.v2 as imageio

image_file="https://media.cheggcdn.com/media/216/21621ee5-e80f-47f3-9145-513f2229b390/phploeBuh.png"

mask_image = imageio.imread(image_file)
```

```python
from wordcloud import WordCloud
mask= mask_image

word_cloud = WordCloud(ranks_only= frequency, background_color = 'white', mask=mask).generate(data)

plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
