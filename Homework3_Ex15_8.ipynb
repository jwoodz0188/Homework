```python
# import needed libaries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
#load the iris data
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# determine the optimal k value(k range fromm 1 to 31)

k_range = range(1,31)
k_scores = []
for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  scores = cross_val_score(knn, X, y, cv=10)
  k_scores.append(scores.mean())
```

```python
#plot 'Value of K for KNN' vs. 'Cross-Validated Accuracy'
plt.plot(k_range, k_scores )
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validate Accuracy')
plt.show()
```

```python
#Using GridSearchCV to confirm K=13 has the optimal parameter
param_grid= {'n_neighbors' : k_range}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=10, scoring='accuracy')
grid.fit(X, y)
print("The optimal parameter for k is: ", grid.best_params_['n_neighbors'])
```
